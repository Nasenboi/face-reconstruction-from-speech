```python
SAMPLE_RATE     = 16_000
AUDIO_LENGTH_S  = 3

NUM_SAMPLES = SAMPLE_RATE * AUDIO_LENGTH_S
```

```python
import librosa
def load_audio(path: str):
    y, sr = librosa.load(path, mono=True, sr=SAMPLE_RATE)
    if len(y) > NUM_SAMPLES:
        y = librosa.effects.trim(y)[0]
    if len(y) > NUM_SAMPLES:
        y = y[:NUM_SAMPLES]
    elif len(y) < NUM_SAMPLES:
        padding = NUM_SAMPLES - len(y)
        pad_left = padding // 2
        pad_right = padding - pad_left
        y = np.pad(y, (pad_left, pad_right), mode='constant')
    return y
```

```python
MAX_VAL_FILES = 1000

X_val = np.array([load_audio(path) for path in val_df["audio_path"].head(MAX_VAL_FILES)])
X_val = X_val.reshape(-1, NUM_SAMPLES, 1)

y_val = val_df[ID_COLUMN_NAMES].head(MAX_VAL_FILES).values
```

```python
MAX_TRAIN_FILES_PER_BATCH = 4000
MAX_BATCH = (len(train_df) + MAX_TRAIN_FILES_PER_BATCH - 1) // MAX_TRAIN_FILES_PER_BATCH  # Ceiling division

def get_train_batch(i=0):
    start = i * MAX_TRAIN_FILES_PER_BATCH
    end = min(start + MAX_TRAIN_FILES_PER_BATCH, len(train_df))
    
    # Use iloc to get the specific slice of the dataframe
    batch_df = train_df.iloc[start:end]
    
    X_train = np.array([load_audio(path) for path in batch_df["audio_path"]])
    X_train = X_train.reshape(-1, NUM_SAMPLES, 1)
    
    y_train = batch_df[ID_COLUMN_NAMES].values
    
    return X_train, y_train
```

```python
import time

VERSION = "1.0"
CREATION_TIME = time.strftime("%d.%m.%H.%M.%S", time.localtime())
OUTPUT_SIZE = 80
AUDIO_MAXLEN = 246000
LABEL_MAXLEN = 256
BATCH_SIZE = 2
```

```python
# https://www.tensorflow.org/hub/tutorials/wav2vec2_saved_model_finetuning
```

```python
import os
os.environ['TF_USE_LEGACY_KERAS']='1'
import tensorflow as tf
import tensorflow_hub as hub

pretrained_layer = hub.KerasLayer("https://tfhub.dev/vasudevgupta7/wav2vec2/1", trainable=False)

inputs = tf.keras.Input(shape=(NUM_SAMPLES,))
reshaped_inputs = tf.keras.layers.Reshape((NUM_SAMPLES, 1))(inputs)
padding_needed = AUDIO_MAXLEN - NUM_SAMPLES
padded_inputs = tf.keras.layers.ZeroPadding1D(padding=padding_needed // 2)(reshaped_inputs)
padded_inputs = tf.keras.layers.Reshape((AUDIO_MAXLEN,))(padded_inputs)
hidden_states = pretrained_layer(padded_inputs)

if isinstance(hidden_states, dict):
    # If it returns a dictionary, extract the relevant tensor
    hidden_states = hidden_states['last_hidden_state']
elif isinstance(hidden_states, (list, tuple)):
    # If it returns multiple outputs, use the last one or aggregate
    hidden_states = hidden_states[-1]

# Global average pooling to handle variable sequence length
x = tf.keras.layers.GlobalAveragePooling1D()(hidden_states)
x = tf.keras.layers.Dense(512, activation='relu')(hidden_states)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.3)(x)
outputs = tf.keras.layers.Dense(OUTPUT_SIZE, activation='linear')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)

optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(
    loss="mean_squared_error",
    optimizer=optimizer,
    metrics=["mae"]
)

@tf.function(jit_compile=True)
def forward(speech):
    return model(speech, training=True)

print(model.summary())
```

```python
EPOCHS = 5
MAX_BATCHES = 1
BATCHES = min(MAX_BATCHES, MAX_BATCH) 

for i in range(BATCHES):
    X_train, y_train = get_train_batch(i)
    history = model.fit(
        X_train,
        y_train,
        batch_size=12,
        epochs=EPOCHS,
        validation_data=(X_val, y_val),
        verbose=1,
        shuffle=True
    )
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

test_predictions = model.predict(X_train, verbose=1)

y_train_0, y_train_0_pred = y_train[:,0], test_predictions[:, 0]

print(f"Mean Absolute Error: {mean_absolute_error(y_train_0, y_train_0_pred):.4f}")
print(f"Mean Squared Error : {mean_squared_error(y_train_0, y_train_0_0_pred):.4f}")
print(f"R2 Score           : {r2_score(y_train_0, y_train_0_pred):.4f}")
```
